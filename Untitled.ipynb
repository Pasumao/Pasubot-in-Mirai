```python
import requests
from datetime import datetime
import time
import threading
import re
import json
from math import sqrt,pi
from numpy import exp
import pandas as pd
import os
import numpy as np

class Danmu():
    def __init__(self,r,path,uid):
        self.url = 'https://api.live.bilibili.com/xlive/web-room/v1/dM/gethistory';
        self.headers = {
            'Host':'api.live.bilibili.com',
            'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:81.0) Gecko/20100101 Firefox/81.0',
        }
        self.data = {
            'roomid':'{}'.format(r),
            'csrf_token':'',
            'csrf':'',
            'visit_id':'',
        }
        t=str(datetime.today())
        self.pathh=path+'{}\\danmu\\'.format(uid)
        max=0
        nb=0
        for i in os.listdir(self.pathh):
            print(i,i[0])
            if int(i[0])>int(max):
                max=i[0]
        if int(max)==9:
            os.remove(self.pathh+os.listdir(path)[-1])
        for i in os.listdir(self.pathh):
            os.rename(self.pathh+i,self.pathh+str(int(i[0])+1)+i[1:])
        self.filename=str(nb)+r+t[5:7]+'月'+t[8:10]+'日'+t[11:13]+'时'+t[14:16]+'分'
        self.danmulist=[]
        self.ns=0
        
    def get_danmu(self):
        try:
            html = requests.post(url=self.url,headers=self.headers,data=self.data).json();
            with open('{}{}.txt'.format(self.pathh,self.filename),mode='a+',encoding='utf-8') as f:
                self.ns=0
                for content in html['data']['room']:
                    # 获取发言
                    text = content['text']
                    # 获取发言时间
                    timeline = content['timeline']
                    nickname=content['nickname']
                    # 记录发言
                    msg = str(timeline)+' '+nickname+': '+text
                    # 判断对应消息是否存在于日志，如果和最后一条相同则打印并保存
                    if msg not in self.danmulist:
                        self.danmulist.append(msg)
                        # 打印消息
                        print (msg)
                        self.ns=self.ns+1
                        # 保存日志
                        f.write(msg+'\n')
                    # 清空变量缓存
                    nickname = ''
                    text = ''
                    timeline = ''
                    msg = ''
        except:
            print('出错了')
            time.sleep(60)
            pass
        
def dm(r,path,uid):
    global live_status,asd
    bDanmu = Danmu(r,path,uid)
    asd=1
    slptime=0.5
    timelist=[]
    while live_status==1:
        time.sleep(slptime)
        bDanmu.get_danmu();
        if len(timelist)<2:
            timelist.append(bDanmu.ns)
        elif len(timelist)==2:
            timelist[0]=(timelist[0]*timelist[1])/2
            timelist[1]=bDanmu.ns
            cha=timelist[1]-timelist[0]
            if cha==0:
                slptime+=0.3
            else:
                slptime=slptime-cha*0.2
            if slptime<=0.5:
                slptime=0.5
    asd=0

def dmlz(roomid,path,uid):
    global live_status,asd
    u = 30
    sig = 10
    asd=0
    tbool=False
    print(roomid,path,uid)
    url=r'https://api.live.bilibili.com/xlive/web-room/v1/index/getRoomPlayInfo?room_id={}&play_url=1&mask=1&qn=0&platform=web'.format(roomid)
    headers = {
    'Host':'api.live.bilibili.com',
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:81.0) Gecko/20100101 Firefox/81.0',
    }
    html=requests.get(url,headers=headers).json()
    abs_path=path
    while 1:
        datadf=pd.read_excel(abs_path+'dylist.xls')
        for i in datadf.values.tolist():
            tbool=tbool or uid==str(i[2])
        if tbool==False:
            print('tbool')
            return 0
        if html['data']['play_url']==None:
            live_status=0
            time.sleep(10)
        else:
            if asd==0:
                live_status=1
                tdm = threading.Thread(target=dm,args=(roomid,path,uid,))
                tdm.daemon = True
                tdm.start()
        html=requests.get(url,headers=headers).json()
        x=int(datetime.now().strftime('%M'))
        slpt = ((np.exp(-(x - u) ** 2 / (2 * sig ** 2)) / (sqrt(2 * pi) * sig))/0.045)*290+50
        time.sleep(slpt)

if __name__ == "__main__":
    names = locals()
    abs_path=os.path.abspath('.')+"\\awesome_bot\\plugins\\v\\"
    datadf=pd.read_excel(abs_path+'dylist.xls')
    for i in datadf.values.tolist():
        print(i)
        names['t' + str(i[2]) ] = threading.Thread(target=dmlz,args=(str(i[3]),abs_path,str(i[2]),))
        names['t' + str(i[2]) ].daemon = True
        names['t' + str(i[2]) ].start()
        print()
    while 1:
        time.sleep(100)
        datadf=pd.read_excel(abs_path+'dylist.xls')
        for i in datadf.values.tolist():
            if not names['t' + str(i[2]) ].is_alive():
                print(i+'重启')
                names['t' + str(i[2]) ] = threading.Thread(target=dmlz,args=(str(i[3]),abs_path,str(i[2]),))
                names['t' + str(i[2]) ].daemon = True
                names['t' + str(i[2]) ].start()

```
